# Initiate API Key


```python
from openai import OpenAI
import os

%load_ext dotenv
#%reload_ext dotenv
%dotenv

client = OpenAI(
    api_key=os.environ.get("API_TOKEN"),
)
```

## AI Tool Selector Assistants API experiment

```python
import time

assistant = client.beta.assistants.retrieve("asst_KlfI4erbghKGz2aJtC9ZHnQh")

thread = client.beta.threads.create()


message = client.beta.threads.messages.create(
    thread_id=thread.id,
    role="user",
    content="I need access to multiple LLM's via API. What tools does Harvard provide for access?",
    #file_ids=[file.id]
)

run = client.beta.threads.runs.create(
    thread_id=thread.id,
    assistant_id=assistant.id,
)

status = None
while status != "completed":
    run = client.beta.threads.runs.retrieve(
        thread_id=thread.id,
        run_id=run.id
    )
    status = run.status
    if status != "completed":
        time.sleep(2)



messages = client.beta.threads.messages.list(
    thread_id=thread.id
)

assistant_response = messages.data[0].content[0].text.value
print(assistant_response)
```
